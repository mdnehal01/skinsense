```python

# Create the DataFrame
df = pd.DataFrame({
    'Age': age_values,
    'Gender': gender_values,
    'Concerns': concerns_values,
    'DarkPatches': dark_patches_values,
    'DarkSpots': dark_spots_values,
    'Tanned': tanned_values,
    'FineLine': fine_line_values,
    'Wrinkles': wrinkles_values,
    'Beard': beard_values,
    'SensitiveSkin': sensitive_skin,
    'SkinIssues': skin_issues
})

# Display the first few rows of the dataset
print(df.head())

# Save the dataset to a CSV file
df.to_excel('skinMain2.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
dataset = pd.read_excel('skinMain.xlsx')
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

dataset['Gender'] = le.fit_transform(dataset['Gender'])
dataset['Concerns'] = le.fit_transform(dataset['Concerns'])
dataset['DarkPatches'] = le.fit_transform(dataset['DarkPatches'])
dataset['DarkSpots'] = le.fit_transform(dataset['DarkSpots'])
dataset['Tanned'] = le.fit_transform(dataset['Tanned'])
dataset['FineLine'] = le.fit_transform(dataset['FineLine'])
dataset['Wrinkles'] = le.fit_transform(dataset['Wrinkles'])
dataset['Beard'] = le.fit_transform(dataset['Beard'])
```

```python
x = dataset.iloc[:, 0:9].values
x
```

```python
y1 = dataset.iloc[:, 9:10].values
y1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x, y1, test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

model1 = LogisticRegression()
model1.fit(x_train, y_train)
# Predict Y values based on X
y_pred = model1.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test, y_pred)

print(cm)

print(ac)
```
