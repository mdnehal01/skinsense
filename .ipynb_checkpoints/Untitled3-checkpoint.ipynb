```python
import pandas as pd
import random

# Define the number of rows in the dataset
num_rows = 1000

# Define possible values for each column
age_values = [random.randint(15, 65) for _ in range(num_rows)]
gender_values = [random.choice(['Male', 'Female']) for _ in range(num_rows)]

acne_appear = random.choice(['Once a month', 'Once in 3 months', 'Frequently', 'Rarely'] for _ in range(num_rows))
comedones_white_black_tinypimple = random.choice(['More than 5', 'Less than 5', 'None'] for _ in range(num_rows))
pimples = random.choice(['More than 5', 'Less than 5', 'None'] for _ in range(num_rows))
puss_filles_pimples = random.choice(['More than 5', '2-5', 'None'] for _ in range(num_rows))
pits_depression = random.choice(['More than 5', '2-5', 'None'] for _ in range(num_rows))

fine_line_values = [random.choice(['Deep', 'Few', 'No']) for _ in range(num_rows)]
wrinkles_values = [random.choice(['Prominant', 'Few','No']) for _ in range(num_rows)]
skin_react = [random.choice(['Oftern', 'Never']) for _ in range(num_rows)]
forhead_oil = [random.choice(['Oily', 'Dry', 'Normal']) for _ in range(num_rows)]
nose_oil = [random.choice(['Oily', 'Dry', 'Normal']) for _ in range(num_rows)]
cheeks_oil = [random.choice(['Oily', 'Dry', 'Normal']) for _ in range(num_rows)]
skin_feel = [random.choice(['Oily', 'Almost greasy', 'Normal']) for _ in range(num_rows)]

aging = []
oily = []
sensitive = []
problem = []

for i in range(num_rows):
    if acne_appear[i]=='Frequently':
        sensitive.append(1)
        problem.append('Inflammation: infection within the clogged pores')
    
    elif acne_appear[i]=='Once a month' and puss_filles_pimples[i]=='More than 5':
        sensitive.append(1)
        problem.append('Diet and Lifestyle: Follow healthy Diet')
    
    elif pits_depression[i]=='More than 5' and (puss_filles_pimples[i]=='2-5' or pimples[i]=='Less than 5' or pimples[i]=='More than 5'):
        problem.append('Scarring: Severe or untreated acne can lead to permanent scarring.')

    
    elif age_values[i] >= 35 and wrinkles_values[i] == 'Prominant':
        aging.append(1)
    
    elif age_values[i] >=35 and fine_line_values[i] == 'Deep':
        aging.append(1)
        
    elif age_values[i] >=35 and wrinkles_values[i] == 'Prominent' and fine_line_values[i] == 'Deep':
        aging.append(1)
        sensitive.append(1)
    
    elif nose_oil[i] == 'Oily' and forhead_oil[i] == 'Oily' and cheeks_oil[i] == 'Oily':
        oily.append(1)
        sensitive.append(1)
        problem.append('Oily Skin')

    elif skin_feel[i]=='Almost greasy':
        oily.append(1)
    
    elif (nose_oil[i] == 'Oily' or forhead_oil[i] == 'Oily' or cheeks_oil[i] == 'Oily') and (skin_feel[i]=='Almost greasy' or skin_feel[i]=='Oily'):
        oily.append(1)
    
    else:
        oily.append(0)
        aging.append(0)
        problem.append('Normal Skin')

        
df = pd.DataFrame({
    'Age': age_values,
    'Gender': gender_values,
    'Acne Appear': acne_appear,
    'White black Tiny pimples': comedones_white_black_tinypimple,
    'Pimples': pimples,
    'Puss Filled Pimples': puss_filles_pimples,
    'Past pimples spots': pits_depression,
    
    'FineLine': fine_line_values,
    'Wrinkles': wrinkles_values,
    'SkinReact': skin_react,
    'Forehead Oil': forhead_oil,
    'Nose Oil': nose_oil,
    'Cheeks Oil': cheeks_oil,
    'Skin Feels': skin_feel,
    
    'Aging': aging,
    'Oily Skin': oily,
    'Problem':problem
    
})

df.to_excel('Acne.xlsx', index=False)
```
