```python
import pandas as pd
import random

# Define the number of rows in the dataset
num_rows = 1000

# Define possible values for each column
age_values = [random.randint(15, 65) for _ in range(num_rows)]
gender_values = [random.choice(['Male', 'Female']) for _ in range(num_rows)]
concerns_values = [random.choice(['Acne', 'Tan', 'Both','No issues']) for _ in range(num_rows)]
        
fine_line_values = [random.choice(['Deep', 'Few', 'No']) for _ in range(num_rows)]
wrinkles_values = [random.choice(['Prominant', 'Few','No']) for _ in range(num_rows)]
skin_react = [random.choice(['Oftern', 'Never']) for _ in range(num_rows)]
forhead_oil = [random.choice(['Oily', 'Dry', 'Normal']) for _ in range(num_rows)]
nose_oil = [random.choice(['Oily', 'Dry', 'Normal']) for _ in range(num_rows)]
cheeks_oil = [random.choice(['Oily', 'Dry', 'Normal']) for _ in range(num_rows)]
skin_feel = [random.choice(['Oily', 'Almost greasy', 'Normal']) for _ in range(num_rows)]


# acne_appear = []
# comedones_white_black_tinypimple = []
# pimples = []
# puss_filles_pimples = []
# pits_depression = []
# dark_patches_values = []
# dark_spots_values = []
# tanned_values = []
# sunscreen = []

acne_appear = ["" for _ in range(num_rows)]
comedones_white_black_tinypimple = ["" for _ in range(num_rows)]
pimples = ["" for _ in range(num_rows)]
puss_filles_pimples = ["" for _ in range(num_rows)]
pits_depression = ["" for _ in range(num_rows)]
dark_patches_values = ["" for _ in range(num_rows)]
dark_spots_values = ["" for _ in range(num_rows)]
tanned_values = ["" for _ in range(num_rows)]
sunscreen = ["" for _ in range(num_rows)]



for i in range(num_rows):
    if concerns_values[i] == 'Acne':
        acne_appear = [random.choice(['Once a month', 'Once in 3 month', 'Frequently', 'rarely']) for _ in range(num_rows)]
        comedones_white_black_tinypimple = [random.choice(['More than 5', 'Less than 5', 'None']) for _ in range(num_rows)]
        pimples = [random.choice(['More than 5', 'Less than 5', 'None']) for _ in range(num_rows)]
        puss_filles_pimples = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        pits_depression = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        
        dark_patches_values[i] = 'NO'
        dark_spots_values[i] = 'NO'
        tanned_values[i] = 'NO'
        sunscreen[i] = 'NO'
        
    elif concerns_values[i] == 'Tan':
        acne_appear[i] = 'NO'
        comedones_white_black_tinypimple[i] = 'NO'
        pimples[i] = 'NO'
        puss_filles_pimples[i] = 'NO'
        pits_depression[i] = 'NO'
        
        dark_patches_values = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        dark_spots_values = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        tanned_values = [random.choice(['Single Shade darker', 'Multiple Shade darker', 'None']) for _ in range(num_rows)]
        sunscreen = [random.choice(['Frequently/Daily', 'Rarely/Never']) for _ in range(num_rows)]
    
    elif concerns_values[i] == 'Both':
        acne_appear = [random.choice(['Once a month', 'Once in 3 month', 'Frequently', 'rarely']) for _ in range(num_rows)]
        comedones_white_black_tinypimple = [random.choice(['More than 5', 'Less than 5', 'None']) for _ in range(num_rows)]
        pimples = [random.choice(['More than 5', 'Less than 5', 'None']) for _ in range(num_rows)]
        puss_filles_pimples = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        pits_depression = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        
        dark_patches_values = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        dark_spots_values = [random.choice(['More than 5', '2-5', 'None']) for _ in range(num_rows)]
        tanned_values = [random.choice(['Single Shade darker', 'Multiple Shade darker', 'None']) for _ in range(num_rows)]
        sunscreen = [random.choice(['Frequently/Daily', 'Rarely/Never']) for _ in range(num_rows)]
    
    elif concerns_values[i] == 'No issues':
        acne_appear[i] = 'NO'
        comedones_white_black_tinypimple[i] = 'NO'
        pimples[i] = 'NO'
        puss_filles_pimples[i] = 'NO'
        pits_depression[i] = 'NO'
        
        dark_patches_values[i] = 'NO'
        dark_spots_values[i] = 'NO'
        tanned_values[i] = 'NO'
        sunscreen[i] = 'NO'
```

```python
df = pd.DataFrame({
    'Age': age_values,
    'Gender': gender_values,
    'Concerns': concerns_values,
    
    'Acne Appear': acne_appear,
    'White black Tiny pimples': comedones_white_black_tinypimple,
    'Pimples': pimples,
    'Puss Filled Pimples': puss_filles_pimples,
    'Past pimples spots': pits_depression,
    
    'DarkPatches': dark_patches_values,
    'DarkSpots': dark_spots_values,
    'Tanned': tanned_values,
    'Sunscreen': sunscreen,
    
    'FineLine': fine_line_values,
    'Wrinkles': wrinkles_values,
    'SkinReact': skin_react,
    'Forehead Oil': forhead_oil,
    'Nose Oil': nose_oil,
    'Cheeks Oil': cheeks_oil,
    'Skin Feels': skin_feel
    
})

# Display the first few rows of the dataset
print(df.head(1000))
```
