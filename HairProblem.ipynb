```python
import pandas as pd
import random
import numpy as np
import itertools



# Define the number of rows in the dataset
num_rows = 1000

# Define possible values for each column
age_values = [random.randint(15, 70) for _ in range(num_rows)]
gender_values = [random.choice(['Male', 'Female']) for _ in range(num_rows)]
hair_type = [random.choice(['Straight', 'Wavy', 'Curvy']) for _ in range(num_rows)]
thickness = [random.choice(['Thin', 'Normal', 'Thick']) for _ in range(num_rows)]
hair_volume = [random.choice(['Low', 'Moderate', 'Volumized']) for _ in range(num_rows)]
scalp_feel = [random.choice(['Oily', 'Very Oily', 'Dry', 'Very Dry', 'Neither']) for _ in range(num_rows)]
hair_loss = [random.choice(['Few', 'Dozens', 'More']) for _ in range(num_rows)]
medications = [random.choice(['No', 'Not Sure', 'Yes']) for _ in range(num_rows)]
dandruff = [random.choice(['Yes', 'No']) for _ in range(num_rows)]
about_hair = [random.choice(['Very Exposed', 'Less Exposed', 'Not Exposed']) for _ in range(num_rows)]
```

```python
hair_damage = []
scalp_damage = []
product_should_applied = []

for i in range(num_rows):
    # Calculate hair damage based on multiple factors
    if hair_loss[i] == 'Few':
        if medications[i] == 'No' and hair_type[i] in ['Straight', 'Wavy']:
            hair_damage.append('No/Mild')
        else:
            hair_damage.append('Mild/High')
    else:
        hair_damage.append('Very High')

    # Calculate scalp damage based on multiple factors
    if dandruff[i] == 'Yes':
        if scalp_feel[i] in ['Very Oily', 'Dry'] and hair_type[i] != 'Curvy':
            scalp_damage.append('Very High')
        else:
            scalp_damage.append('Mild')
    else:
        scalp_damage.append('No')
     
 # Recommend products based on hair and scalp condition
    if hair_damage[i] == 'Very High' and scalp_damage[i] == 'Very High':
        product_should_applied.append('Specialized Hair and Scalp Treatment Products')
    elif hair_damage[i] == 'Very High':
        product_should_applied.append('Specialized Hair Treatment Products')
    elif scalp_damage[i] == 'Very High':
        product_should_applied.append('Specialized Scalp Treatment Products')
    else:
        product_should_applied.append('Regular Hair and Scalp Care Products')

        
```

```python
df = pd.DataFrame({
    'Age': age_values,
    'Gender': gender_values,
    'Hair Type': hair_type,
    'Thickness': thickness,
    'Volume': hair_volume,
    'Scalp': scalp_feel,
    'Hair Loss': hair_loss,
    'Medications': medications,
    'Dandruff': dandruff,
    'About Exposure': about_hair,
    'Hair Damage': hair_damage,
    'Scalp Damage': scalp_damage,
    'Product Applied': product_should_applied
})

# Display the first few rows of the dataset
print(df.head())

# Save the dataset to a CSV file
df.to_excel('HairMain.xlsx', index=False)
```

```python
data = pd.read_excel('HairMain.xlsx')
data
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
data['Dandruff'] = le.fit_transform(data['Dandruff'])
df = pd.get_dummies(data, columns=['Hair Type', 'Thickness', 'Volume', 'Scalp', 'Hair Loss', 'Medications', 'About Exposure', 'Hair Damage', 'Scalp Damage', 'Product Applied'], prefix=['Hair Type', 'Thickness', 'Volume', 'Scalp', 'Hair Loss', 'Medications', 'About Exposure',  'Hair Damage', 'Scalp Damage', 'Product Applied'])
df
```

```python
label_encoder = LabelEncoder()

# Apply Label Encoding to all columns in the DataFrame
df_encoded = df.apply(label_encoder.fit_transform)
df_encoded
```

```python
x = df_encoded.iloc[:, 0:26].values
x[0]
```

```python
y = df_encoded.iloc[:, 26:].values
y
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load or prepare your dataset
# X contains input features
# y contains multivariate output variables

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create a Multi-Output Decision Trees model
base_model = DecisionTreeRegressor()
model = MultiOutputRegressor(base_model)

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

print(y_pred)
```
